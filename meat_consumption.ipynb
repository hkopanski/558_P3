```r
library(tidyverse)
```

```r
df_meat <- read_csv("meat_consumption_worldwide.csv")
df_meat2 <- read_csv("DP_LIVE_19072021012255894.csv")
```

```r
str(df_meat)
summary(df_meat)
head(df_meat)

str(df_meat2)
summary(df_meat2)
head(df_meat2)
```

```r
df_meat <- df_meat %>% rename(YEAR = TIME, UNITS = MEASURE, MEAT = SUBJECT)
```

```r
df_meat %>% select(LOCATION) %>% unique() %>% nrow()
```

```r
df_c_codes <- read_csv("region_code.csv", col_names = c("REGION", "LOCATION", "CODE"), skip = 1)
```

```r
head(df_c_codes)
```

```r
df_meat <- df_meat %>% left_join(df_c_codes, by = "LOCATION")
print(summary(df_meat))
print(df_meat)
```

```r
#df_meat %>% select(LOCATION, REGION) %>% filter(is.na(REGION)) %>% unique() %>% print()
na_regions <- df_meat %>% select(LOCATION, REGION) %>% filter(is.na(REGION)) %>% unique() #%>% select(LOCATION) %>% pull()
#na_regions <- as.vector(na_regions)
print(na_regions)

missing_info <- c("World", "Subsaharan Africa", "Developed Countries", "Brazil/Russia/India/China/South Africa", "EU including UK")

for(i in 1:5){
    na_regions[i, 2] = missing_info[i]
}
```

```r
print(na_regions)
```

```r
df_meat3 <- df_meat #%>% left_join(na_regions, by = "LOCATION")
print(df_meat3)
df_meat3 %>% filter(is.na(REGION)) %>% print()
```

```r
for(i in 1:nrow(na_regions)){
    for(j in 1:nrow(df_meat3)){
        if(df_meat3[j, "LOCATION"] == na_regions[i, "LOCATION"]){
            df_meat3[j, "REGION"] = na_regions[i, "REGION"]
        }
    }
}
df_meat3 %>% print()
```

```r
print((df_meat3))
df_meat3 %>% filter(LOCATION == "BRICS") %>% print()
```

```r
df_meat3 %>% filter(LOCATION == "AUS", MEAT == "BEEF")
```

```r
a <- df_meat3 %>% filter(MEAT == "BEEF", LOCATION == "WLD", UNITS == "KG_CAP") %>% select(Value) %>% colMeans()
b <- df_meat3 %>% filter(MEAT == "BEEF", LOCATION == "WLD", UNITS == "THND_TONNE") %>% select(Value) %>% colMeans()

coeff_fac <- b/a

df_temp <- df_meat3 %>% filter(MEAT == "BEEF", LOCATION == "WLD")

df_meat3 %>% filter(MEAT == "BEEF", LOCATION == "WLD", UNITS == "KG_CAP") %>% ggplot(aes(x = YEAR)) + geom_line(aes(y = Value))
df_meat3 %>% filter(MEAT == "BEEF", LOCATION == "WLD", UNITS == "THND_TONNE") %>% ggplot(aes(x = YEAR)) + geom_line(aes(y = Value))
```

```r
df_temp <- df_meat3 %>% filter(MEAT == "SHEEP", LOCATION == "WLD")
#df_temp

#names(df_temp)

df_temp <- spread(df_temp, UNITS, Value)
print(df_temp)

coeff_value <- mean(df_temp$THND_TONNE)/mean(df_temp$KG_CAP)

df_temp %>% ggplot(aes(x = YEAR)) + geom_line(aes(y = KG_CAP), col = 'red') + geom_point(aes(y = KG_CAP), col = 'red') + 
    geom_line(aes(y = THND_TONNE / coeff_value), col = 'blue') +
    geom_point(aes(y = THND_TONNE / coeff_value), col = 'blue') + 
    scale_y_continuous(name = "Kilograms per Capita",
                   sec.axis = sec_axis(~.*coeff_value, name="Thousands of Tonnes")) + 
    labs(title = paste("Consumption of", unique(df_temp$MEAT), "from", unique(df_temp$REGION)))
```
